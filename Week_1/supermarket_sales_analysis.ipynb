```python
# import necessary libraries
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
# lets visualize the Region wise order count bar plot 
sns.barplot(x='Region', y='Order_ID', data=df, estimator=len)
```

```python
# lets visualize the Country wise deals count bar plot
sns.barplot(x='Country', y='Order_ID', data=df, estimator=len)
```

```python
# lets check the Sales Rep wise performance bar plot
sns.barplot(x='Order_ID', y='Sales_Rep', data=df, estimator=len)
```

```python
# lets check the Region vs Revenue bar plot
sns.barplot(x='Region', y='Revenue', data=df, estimator=sum)
```

```python
# lets check the Product Category vs Revenue bar chart
sns.barplot(x='Product_Category', y='Revenue', data=df, estimator=sum)
```

```python
# lets check the Revenue vs Target scatter plot
sns.scatterplot(x='Revenue', y='Target', data=df, hue='Stage')


```

```python
sns.scatterplot(x='Revenue', y='Target', data=df, hue='Region')
```

```python
# lets check the Order_Date vs Revenue line chart
sns.lineplot(x='Order_Date', y='Revenue', data=df)
```

```python
# lets check the Pie chart: Won vs Lost vs Opportunity
stage_counts = df['Stage'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(stage_counts, labels=stage_counts.index, autopct='%1.1f%%', startangle=140,
		colors=sns.color_palette('pastel'))
plt.axis('equal')
plt.show()
```

```python
# lets check the Histograms of numerical columns
def make_hist(col_name):
    plt.figure(figsize=(7,5))
    sns.histplot(data=df, x=col_name, kde=True, palette='Set2')
    plt.title(f"Histogram of {col_name}")
    plt.show()

make_hist('Revenue')
make_hist('Target')
make_hist('Units_Sold')
make_hist('Deal_Size')
```
